```python
#problem 1
import numpy as np
import matplotlib.pyplot as plt
from math import e
def f(x):
  return -4*x*(x**2-1)
def euler(b,f) :         #defining euler method
  h=0.01
  y=b
  y+=h*f
  return y
P1=np.array([])
P1=np.append(P1,0.1)
X1=np.array([])
X1=np.append(X1,1)
for i in range (10000):
  p=euler(P1[i],f(X1[i]))
  P1=np.append(P1,p)
  x=euler(X1[i],P1[i])
  X1=np.append(X1,x)
plt.plot(X1,P1)
plt.show()
plt.savefig("1.pdf")
P2=np.array([])
P2=np.append(P2,0.1)
X2=np.array([])
X2=np.append(X2,-1)
for i in range (10000):
  p=euler(P2[i],f(X2[i]))
  P2=np.append(P2,p)
  x=euler(X2[i],P2[i])
  X2=np.append(X2,x)
plt.plot(X2,P2)
plt.show()
plt.savefig("2.pdf")
P3=np.array([])
P3=np.append(P3,10)
X3=np.array([])
X3=np.append(X3,1)
for i in range (10000):
  p=euler(P3[i],f(X3[i]))
  P3=np.append(P3,p)
  x=euler(X3[i],P3[i])
  X3=np.append(X3,x)
plt.plot(X3,P3)
plt.show()
plt.savefig("3.pdf")
```
