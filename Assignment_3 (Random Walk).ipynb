```python
import matplotlib.pyplot as plt
from numpy import random
import numpy as np
#fORWARD=1
#bACKWARD=-1
dir=[1,-1]
x=0
pos=[x]
I=[]
random.seed(5)
for i in range(3600):
  I.append(i)
  y=random.choice(dir)
  if(y==1):
    x+=1
  elif (y==-1):
    x+=-1
  pos.append(x)
pos.pop(-1)
plt.plot(I,pos)
plt.xlabel("time(seconds)")
plt.ylabel("position")
plt.title("random walk in 1D")

#plt.show()
plt.savefig("rd1.pdf")
```

```python
import matplotlib.pyplot as plt
from numpy import random
import numpy as np                              #mean and rms for increasing time
I=[]
D=[]
M=[]
RMS=[]
steps=1000
s=[]
for i in range(1,101):
  random.seed(i)
  dir=[1,-1]
  y=0
  pos=[y]
  d_ment=0
  s.append(i)
  for j in range(steps*i):



    x=random.choice(dir)
    if(x==1):
     y+=1
    elif (x==-1):
      y+=-1
      pos.append(y)
  d_ment+=pos[-1]

  D.append(d_ment)
  mean=np.mean(D)
  M.append(mean)
  #print("mean:",mean)
  sum=0
  for i in range(len(D)):
    sum+=D[i]**2
  rms=(sum/100) **(0.5)
  RMS.append(rms)

  #print("rms value:",rms)
plt.plot(s,M,color="r")
plt.title("mean")
plt.xlabel("Xsteps")
plt.ylabel("mean")
plt.savefig("m.pdf")
plt.show()
plt.plot(s,RMS,color="b")
plt.title("rms")
plt.xlabel("Xsteps")
plt.ylabel("rms")
plt.savefig("rms.pdf")
plt.show()
```
