```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,2*np.pi,0.01)
```

```python
a0=1             #for k>0   omega0>1
omega0=1.2     #0.1<omega0<2
H0=1/(13.58*10**9*365*24*60*60)
a=((a0*omega0)/(2*(omega0-1)))*(1-np.cos(x))
t=(omega0/(2*H0*((omega0-1)**(3/2))))*(x-np.sin(x))
plt.xlabel('\u03B8')

plt.plot(x,a,color="red",label="a vs $\u03B8$")
plt.legend()
plt.show()

plt.plot(x,t,color="purple",label="t vs $\u03B8$")
plt.legend()

plt.show()
```

```python
plt.plot(t,a,color="teal")
plt.xlabel("a")
plt.ylabel("t")
plt.title("a vs t")
```
