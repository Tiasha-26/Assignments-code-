```python
#prob 3
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import special
```

```python
x=np.linspace(0,2.4,1024)
tau=np.linspace(0,0.5,4096)
for i in range(1,len(tau),800):
  sigma=np.exp(-(1+x**2)/tau[i])*scipy.special.iv(1.0/4.0,2*x/tau[i])/(tau[i]*x**(1.0/4.0))
  plt.plot(x,sigma,label=tau[i])
plt.legend(loc="upper right")
plt.ylabel("$\Sigma\pi r_{0}^2/m$")
plt.xlabel(r"$x=r/r_{0}$")
plt.show()
```
